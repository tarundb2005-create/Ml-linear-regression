```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
```

```python
df = pd.read_csv('homeprices.csv')  #usig pandas we read the csv file
```

```python
df.head()    #head is used to see first 5 data to analyse how the data are stored in a dataset
```

```python
plt.scatter(df.area,df.price,marker = '+') #using scatter plot
```

```python
reg = linear_model.LinearRegression()      #using linear regression model

#train dataset using linear regression

reg.fit(df[['area']], df['price'])           #fiting the data in 2d arry and try to avoid the series data
```

```python
reg.predict([[4030]])  #predicting the 4030sq feet price
```

y = m*x+b
m = reg.coef_
b = reg.intercept_

```python
reg.coef_
```

```python
reg.intercept_
```

```python
Pred = reg.coef_*4030+reg.intercept_
Pred
```

```python
if Pred == reg.predict([[4030]]):
  print('the model predicted correct')
else:
  print('the model predicted wrong')
```
