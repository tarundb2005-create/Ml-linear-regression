```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('ads.csv')
df
```

```python
reg= linear_model.LinearRegression()
reg
```

```python
reg.fit(df[['ads']],df[['sales']])
```

```python
reg.predict([[11]])


```

```python
reg.predict([[12]])
```

```python
plt.scatter(df.ads,df.sales,color = 'black',marker = '+')
plt.plot(df.ads,reg.predict(df[['ads']]),color = 'red')
plt.xlabel('ads')
plt.ylabel('sales $')
plt.show()
```
