```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('study.csv')
df
```

```python

reg = linear_model.LinearRegression()
reg
```

```python
reg.score(df[['Hours_Studied']], df[['Exam_Score']])
```

```python
reg.fit(df[['Hours_Studied']],df[['Exam_Score']])
```

```python
reg.predict([[7]])
```

```python

plt.plot(df[['Hours_Studied']],reg.predict(df[['Hours_Studied']]),color='red',marker = '+')
```

```python
print(reg.intercept_)
print(reg.coef_)
print(reg.predict([[7]]))
```

```python
reg.intercept_ + reg.coef_ * 7
```
